```python
# bounding box analysis
import pandas as pd
import numpy as np 


df = pd.read_csv("https://raw.githubusercontent.com/gulvarol/grocerydataset/master/annotations.csv",
                 names=["img", "x", "y", "w", "h", "class"])
df.head()
```

```python
df["class"].unique()
```

```python
import cv2
id = 1
im = df.img[id]

im_df = df[df["img"]==im]
bbox_ls = []


image = cv2.imread("ShelfImages/train/"+im)
print(image.shape)
for i in range(len(im_df["img"])):
  [x,y,w,h] = im_df.x[i], im_df.y[i], im_df.w[i], im_df.h[i]
  cv2.rectangle(image,(x,y),(w,h),(0,255,255),2)

plt.figure(figsize=(15,9))
plt.imshow(image)
plt.show()
```

```python
# bounding box analysis

aspect_ratio_x = df["w"]-df["x"]
aspect_ratio_y = df["h"]-df["y"]

aspect_ratio = aspect_ratio_x/aspect_ratio_y
aspect_ratio.hist()
```
